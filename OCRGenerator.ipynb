```python
from PIL import Image
import pytesseract
import re

# Caminho da imagem
image_path = "/content/drive/MyDrive/.....top-10-cnae-mei-2022.png"

# Abrir a imagem usando o PIL
img = Image.open(image_path)

# Extrair texto da imagem
extracted_text = pytesseract.image_to_string(img, lang='por')

def clean_text(text):
    # Remove espaços em branco no início e no final
    text = text.strip()
    # Substitui múltiplos espaços em branco por um único espaço
    text = re.sub(r'\s+', ' ', text)
    return text

# Supondo que 'extracted_text' seja o texto que você extraiu da imagem
cleaned_text = clean_text(extracted_text)
print(cleaned_text)
```

```python
def segment_into_paragraphs(text):
    # Divide o texto em parágrafos baseado em múltiplas quebras de linha consecutivas
    paragraphs = re.split(r'\n\s*\n', text)
    # Limpa cada parágrafo e remove parágrafos vazios
    paragraphs = [clean_text(p) for p in paragraphs if clean_text(p)]
    return paragraphs

paragraphs = segment_into_paragraphs(extracted_text)

for idx, paragraph in enumerate(paragraphs, 1):
    print(f"\n{paragraph}\n")
```
